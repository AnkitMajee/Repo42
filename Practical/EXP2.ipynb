Generate a random (100x100) matrix
Find for each column (feature)
1. Mean
2. Variance
3. Kurtosis
4. Skewness

Plot 4 graphs of above 4 metrics for each column.

```python
import numpy as np
import pandas as pd
from scipy.stats import kurtosis, skew

# Generate a random (100x100) matrix
matrix = np.random.rand(100, 100)
print(matrix)
```

```python
# Calculate metrics for all columns
means = np.mean(matrix, axis=0)
variances = np.var(matrix, axis=0)
kurtoses = kurtosis(matrix, axis=0)
skewnesses = skew(matrix, axis=0)

df = pd.DataFrame({
    'Column': range(matrix.shape[1]),
    'Mean': means,
    'Variance': variances,
    'Kurtosis': kurtoses,
    'Skewness': skewnesses
})

df
```

```python
import matplotlib.pyplot as plt

# Plot the metrics
plt.figure(figsize=(15, 8))

plt.subplot(2, 2, 1)
plt.plot(df['Column'], df['Mean'], marker='o')
plt.xlabel('Column')
plt.ylabel('Mean')
plt.title('Mean of Each Column')

plt.subplot(2, 2, 2)
plt.plot(df['Column'], df['Variance'], marker='o')
plt.xlabel('Column')
plt.ylabel('Variance')
plt.title('Variance of Each Column')

plt.subplot(2, 2, 3)
plt.plot(df['Column'], df['Kurtosis'], marker='o')
plt.xlabel('Column')
plt.ylabel('Kurtosis')
plt.title('Kurtosis of Each Column')

plt.subplot(2, 2, 4)
plt.plot(df['Column'], df['Skewness'], marker='o')
plt.xlabel('Column')
plt.ylabel('Skewness')
plt.title('Skewness of Each Column')

plt.tight_layout()
plt.show()
```
